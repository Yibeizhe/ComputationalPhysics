```python
import numpy as np
import matplotlib.pyplot as plt

def sawtooth_function(x):
    return x - 2 * np.pi * np.floor(x / (2 * np.pi))

x = np.linspace(-3*np.pi, 3*np.pi, 1000)

x1 = sawtooth_function(x)
y=0.5*(np.pi-x1)

def fouri_rec(x,n):
    yf=0
    for i in range(1,n):
        yf+=np.sin(i*x)/i
    return yf
y_f5=fouri_rec(x1,5)
y_f25=fouri_rec(x1,25)
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Sawtooth Function", color='blue')
plt.plot(x,y_f5,label="n=5")
plt.plot(x,y_f25,label="n=25")
plt.title("Sawtooth Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

要计算积分 $\int \sqrt{4+x^2} \, dx$，我们可以使用三角代换或者直接使用分部积分法。这里我将演示使用三角代换的方法。

令 $x = 2 \tan \theta$，其中 $\theta$ 是新的变量。这将使得 $dx = 2 \sec^2 \theta \, d\theta$。

然后，我们可以将 $\sqrt{4+x^2}$ 转换为三角函数，利用三角恒等式 $\sec^2 \theta = 1 + \tan^2 \theta$：

\begin{align*}
\sqrt{4+x^2} &= \sqrt{4+4\tan^2 \theta} \\
&= \sqrt{4 \sec^2 \theta} \\
&= 2 \sec \theta.
\end{align*}

现在，我们可以将原积分转换为关于 $\theta$ 的积分：

\begin{align*}
\int \sqrt{4+x^2} \, dx &= \int 2 \sec \theta \cdot 2 \sec^2 \theta \, d\theta \\
&= 4 \int \sec^3 \theta \, d\theta.
\end{align*}

这是一个关于 $\sec^3 \theta$ 的积分，我们可以通过分部积分来解决它。分部积分公式为 $\int u \, dv = uv - \int v \, du$，其中 $u$ 和 $dv$ 分别是积分中的两个部分，$du$ 和 $v$ 是它们的导数或积分。

令 $u = \sec \theta$ 和 $dv = \sec^2 \theta \, d\theta$，然后计算出相应的 $du$ 和 $v$：

\begin{align*}
du &= \sec \theta \tan \theta \, d\theta, \\
v &= \int \sec^2 \theta \, d\theta = \tan \theta.
\end{align*}

应用分部积分公式：

\begin{align*}
\int \sec^3 \theta \, d\theta &= \sec \theta \tan \theta - \int \tan \theta \sec \theta \tan \theta \, d\theta \\
&= \sec \theta \tan \theta - \int \sec \theta \tan^2 \theta \, d\theta.
\end{align*}

现在，我们可以使用一个标准的积分公式 $\int \tan^2 \theta \, d\theta = \sec^2 \theta - \theta$：

\begin{align*}
\int \sec^3 \theta \, d\theta &= \sec \theta \tan \theta - (\sec^2 \theta - \theta) \\
&= \theta + \sec \theta \tan \theta - \sec^2 \theta.
\end{align*}

最后，回到原始的变量 $x$：

\begin{align*}
\int \sqrt{4+x^2} \, dx &= 4 \left( \theta + \sec \theta \tan \theta - \sec^2 \theta \right) \\
&= 4 \left( \arctan \frac{x}{2} + \frac{x}{2} \sqrt{1 + \frac{x^2}{4}} - \frac{x^2}{4} \right) + C,
\end{align*}

其中 $C$ 是积分常数。这就是给定积分的结果。请注意，上述推导中使用了三角恒等式和标准的三角函数积分。

要计算定积分 $\int_{0}^{4} \sqrt{4+x^2} \, dx$，我们可以使用之前提到的三角代换方法。首先，我们进行三角代换：

之后，我们可以按照之前的步骤计算出积分：

\begin{align*}

\end{align*}

这就是定积分 $\int_{0}^{4} \sqrt{4+x^2} \, dx$ 在给定积分范围内的结果。
