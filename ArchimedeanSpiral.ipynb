```python
import numpy as np
import matplotlib.pyplot as plt

# 
a = 0.5  # 
b = 0.2  # 

# 
theta = np.linspace(0, 10*np.pi, 1000)

# 
r = a + b*theta

# 
fig1 = plt.figure()
ax1 = fig1.add_subplot(111, projection='polar')
ax1.plot(theta, r)

# 
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
x = r * np.cos(theta)
y = r * np.sin(theta)
ax2.plot(x, y)

plt.show()
```
