```python
import numpy as np
import matplotlib.pyplot as plt

# 定义阿基米德螺旋线的参数
a = 0.5  # 控制线密度的参数
b = 0.2  # 控制线距离中心的距离的参数

# 构造极角数组
theta = np.linspace(0, 10*np.pi, 1000)

# 构造极径数组
r = a + b*theta

# 极坐标系下的绘图
fig1 = plt.figure()
ax1 = fig1.add_subplot(111, projection='polar')
ax1.plot(theta, r)

# 直角坐标系下的绘图
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
x = r * np.cos(theta)
y = r * np.sin(theta)
ax2.plot(x, y)

plt.show()
```
