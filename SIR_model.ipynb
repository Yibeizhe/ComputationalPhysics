# 病毒感染的SIR模型  
## 模型的微分方程
- 易感者（Susceptible，S）,其方程为：$\frac{dS}{dt}=-\beta \cdot S \cdot I$
- 感染者（Infectious，I），
- 康复者（Recovered，R）

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
beta = 0.3  # Infection rate
gamma = 0.1  # Recovery rate

# Initial conditions
S0 = 0.9  # Initial susceptible population
I0 = 0.1  # Initial infected population
R0 = 0.0  # Initial recovered population

# Total time and time step
T = 200  # Total simulation time
dt = 1  # Time step

# Number of time steps
num_steps = int(T / dt)

# Arrays to store S, I, and R values over time
S = np.zeros(num_steps + 1)
I = np.zeros(num_steps + 1)
R = np.zeros(num_steps + 1)

# Initial conditions
S[0] = S0
I[0] = I0
R[0] = R0

# Euler's method for numerical integration
for t in range(num_steps):
    dS = -beta * S[t] * I[t] * dt
    dI = (beta * S[t] * I[t] - gamma * I[t]) * dt
    dR = gamma * I[t] * dt
    S[t + 1] = S[t] + dS
    I[t + 1] = I[t] + dI
    R[t + 1] = R[t] + dR
print('The number of infectious as following: ')
print(I)
# Time values
time = np.linspace(0, T, num_steps + 1)
print('Times is as following: ')
print(time)
# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(time, S, label='Susceptible')
plt.plot(time, I, label='Infected')
plt.plot(time, R, label='Recovered')
plt.xlabel('Time')
plt.ylabel('Fraction of Population')
plt.legend()
plt.title('SIR Model Simulation')
plt.grid(True)
plt.show()
```
