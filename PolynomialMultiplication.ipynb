```python
import numpy as np, matplotlib.pyplot as plt
x=np.linspace(-2,2,99)
y1=x**2-2*x+1
y2=x**2+2*x+1
y3=y1*y2
y4=x**3-6*x**2+11*x-6
y5=2*x**3-9*x**2+12*x+5
plt.plot(x,y1,label="y1")
plt.plot(x,y2,label="y2")
plt.plot(x,y3,label="y3")
plt.plot(x,y4,label="y4")
plt.plot(x,y5,label="y5")
plt.legend()
```
