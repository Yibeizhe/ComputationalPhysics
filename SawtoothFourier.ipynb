锯齿波函数的周期性定义：
<span style="color:red; font-size:20px;">
$$
f(x) = \begin{cases}
\frac{1}{2}(\pi - x) & (0 < x \leq 2\pi) \\
f(x + 2\pi) & (x \notin [0,2\pi])
\end{cases}
$$
</span>
锯齿波函数的傅里叶级数展开：
<span style="color:green; font-size:20px;">
$$
f(x) = \frac{1}{2}(\pi - x) = \sum_{n=1}^{\infty} \frac{\sin nx}{n} \quad (0 < x \leq 2\pi)
$$
</span>

```python
import numpy as np
import matplotlib.pyplot as plt

def sawtooth_function(x):
    return x - 2 * np.pi * np.floor(x / (2 * np.pi))

x = np.linspace(-6*np.pi, 6*np.pi, 1000)

x1 = sawtooth_function(x)
y=0.5*(np.pi-x1)

def fouri_rec(x,n):
    yf=0
    for i in range(1,n):
        yf+=np.sin(i*x)/i
    return yf
y_f5=fouri_rec(x1,5)
y_f25=fouri_rec(x1,25)
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Sawtooth Function", color='blue')
plt.plot(x,y_f5,label="n=2")
plt.plot(x,y_f25,label="n=15")
plt.title("Sawtooth Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```
